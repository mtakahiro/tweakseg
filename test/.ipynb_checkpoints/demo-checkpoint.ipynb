```python
from tweakseg.tweakseg import add_source,save_segmap,extend_segmap    
from astropy.io import fits
import matplotlib.pyplot as plt
```

# Input files
- segmentation map in fits
- photometric catalog file from grizli in fits

```python
file_seg = 'jw01324001001-ir_seg.fits'
file_cat = 'jw01324001001-ir.cat.fits'
```

```python
# Have those prepared;
fd_seg = fits.open(file_seg)[0].data
hd_seg = fits.open(file_seg)[0].header
fd_cat = fits.open(file_cat)[1].data
```

## 1. case where you want to expand your segmap of a specific source

```python
# Show original;
plt.imshow(fd_seg)
plt.xlim(1240,1280)
plt.ylim(1220,1260)
plt.title('Original segment')
```

```python
# Your target id;
id_targ = 3132
# Radius of expansion, in pixel;
radius = 15
# Override surrounding pixel?
override = False #True # If True, surrounding pixels that fall in <r would be changed to id_targ.

fd_seg_new, _ = extend_segmap(fd_seg, id_targ, radius=radius, coords=None, override=override, file_cat=None)
```

```python
plt.imshow(fd_seg_new)
plt.xlim(1240,1280)
plt.ylim(1220,1260)
plt.title('Updated segment')
plt.show()
```

## 2. case where you want to inject a new circular segmap of a specific source

```python
# This empty region
plt.imshow(fd_seg)
plt.xlim(1200,1230)
plt.ylim(1220,1260)
plt.title('Original segment')
plt.show()

```

```python
# where;
coords = [1216, 1244] # x,y

# New id;
id_targ = 30000 # Make sure this won't overlap with existing sources.

# How large;
radius = 5 # in pixel

# Also feed catalog file, so the module will add the source;
file_cat_output = 'test_cat.fits'

fd_seg_new, file_cat_new = extend_segmap(fd_seg_new, id_targ, radius=radius, coords=coords, 
                                         override=True, file_cat=file_cat, file_cat_out=file_cat_output)
```

```python
plt.imshow(fd_seg_new)
plt.xlim(1200,1230)
plt.ylim(1220,1260)
plt.title('Updated segment')
```

## 3. Case where you want to add specific regions by an input polygon files
Make sure to have one region file per object    

```python
# This region
plt.imshow(fd_seg)
plt.xlim(1540,1800)
plt.ylim(1082,1280)
plt.title('Original segment')
plt.show()
```

```python
files_polygon = ['targets_ds9.reg']
ids_targ = [50000]
for file_polygon, id_targ in zip(files_polygon,ids_targ):
    print(file_polygon, id_targ)
    
    fd_seg_new, file_cat_new = extend_segmap(fd_seg_new, id_targ, coords=None, override=True, file_region=file_polygon, 
                                             hd_seg=hd_seg, file_cat=file_cat_new, file_cat_out=file_cat_output)
```

```python
plt.imshow(fd_seg_new)
plt.xlim(1540,1800)
plt.ylim(1082,1280)
plt.title('Updated segment')
```

## 4. Save segmentation map;

```python
file_output = 'test_seg.fits'
save_segmap(fd_seg_new, file_output=file_output, hdr=hd_seg)
```
